```python
import re
import scipy
import string
import pickle
import warnings
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder            
from sklearn.svm import LinearSVC             
from sklearn.naive_bayes import BernoulliNB
from sklearn.ensemble import RandomForestClassifier          
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score    
from astropy.table import Table, Column
from sklearn.model_selection import train_test_split
from sklearn import tree
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import f1_score
from sklearn.ensemble import BaggingClassifier
from prettytable import PrettyTable
from ipywidgets import widgets



sample_data = pd.read_csv("IRIS.csv")

print("\n\nSample Data:")
print("============\n")
pd.set_option("display.max_rows", None, "display.max_columns", None)
print(sample_data)


species=pd.DataFrame({"species":["Iris-setosa","Iris-versicolor","Iris-virginica"]}) 
species_label_encoder = LabelEncoder()
species_label_encoder.fit(np.ravel(species))

sample_data_encoded = sample_data.copy()
original_sample_data = sample_data.copy()

# Transform Output of into Numerical Representation

print("\n\nOutput Attribute After Label Encoding:")
print("========================================\n")
sample_data["encoded_species"] = species_label_encoder.transform(sample_data['species'])
print(sample_data[["species", "encoded_species"]])

# Print Original and Encoded Ouput Sample Data

sample_data_encoded[["sepal_length","sepal_width","petal_length","petal_width","species"]] = sample_data[["sepal_length","sepal_width",
                                                                                                          "petal_length","petal_width","encoded_species"]]
pd.set_option("display.max_rows", None, "display.max_columns", None)
print("\n\nOriginal Sample Data:")
print("=====================\n")
pd.set_option("display.max_rows", None, "display.max_columns", None)
print(original_sample_data)
print("\n\nSample Data after Label Encoding of Output:")
print("===========================================\n")
pd.set_option("display.max_rows", None, "display.max_columns", None)
print(sample_data_encoded)

# Save the Transformed Features into CSV File 

sample_data_encoded.to_csv(r'sample-data-encoded.csv', index = False, header = True)


training_data_encoded, testing_data_encoded = train_test_split( sample_data_encoded , test_size=0.2 , random_state=0 , shuffle = False)
print("")
print("                    Training Data                            ")
print("")
input_vector_train = training_data_encoded.iloc[: , :-1]
print(input_vector_train)
print("species")
output_label_train = training_data_encoded.iloc[: ,-1]
print(output_label_train)
print("")
print("                    Testing Data                            ")
print("")
input_vector_test = testing_data_encoded.iloc[: , :-1]
print(input_vector_test)
print("species")
output_label_test = testing_data_encoded.iloc[: ,-1]
print(output_label_test)
```

```python
logistic_regression=LogisticRegression()
logistic_regression.fit(input_vector_train,output_label_train)
logistic_regression_prediction=logistic_regression.predict(input_vector_test)
predict=logistic_regression_prediction
print(predict)
lr_accuracy=accuracy_score(output_label_test,logistic_regression_prediction)

print(lr_accuracy)
```

```python
Random_Forest_Classifier=RandomForestClassifier()
Random_Forest_Classifier.fit(input_vector_train,output_label_train)
Random_Forest_Classifier_prediction=Random_Forest_Classifier.predict(input_vector_test)
predict=Random_Forest_Classifier_prediction
print(predict)
rfc_accuracy=accuracy_score(output_label_test,Random_Forest_Classifier_prediction)
print(rfc_accuracy)
```

```python
linearsvc=LinearSVC(max_iter=10000)
linearsvc.fit(input_vector_train,output_label_train)
linearsvc_prediction=linearsvc.predict(input_vector_test)
predict=linearsvc_prediction
print(predict)
lsvc_accuracy=accuracy_score(output_label_test,linearsvc_prediction)
print(lsvc_accuracy)
```

```python
bernoullinb=BernoulliNB()
bernoullinb.fit(input_vector_train,output_label_train)
bernoullinb_prediction=bernoullinb.predict(input_vector_test)
predict=bernoullinb_prediction
print(predict)
bnb_accuracy=accuracy_score(output_label_test,bernoullinb_prediction)
print(bnb_accuracy)
```

```python
decisiontree=tree.DecisionTreeClassifier()
decisiontree.fit(input_vector_train,output_label_train)
decisiontree_prediction=decisiontree.predict(input_vector_test)
predict=decisiontree_prediction
print(predict)
dt_accuracy=accuracy_score(output_label_test,decisiontree_prediction)

print(dt_accuracy)
```

```python
kneighbors=KNeighborsClassifier(n_neighbors=7)
kneighbors.fit(input_vector_train,output_label_train)
kneighbors_prediction=kneighbors.predict(input_vector_test)
predict=kneighbors_prediction
print(predict)
knn_accuracy=accuracy_score(output_label_test,kneighbors_prediction)

print(knn_accuracy)
```

```python
gaussian_nb=GaussianNB()
gaussian_nb.fit(input_vector_train,output_label_train)
gaussian_nb_prediction=gaussian_nb.predict(input_vector_test)
predict=gaussian_nb_prediction
print(predict)
gnb_accuracy=accuracy_score(output_label_test,gaussian_nb_prediction)
print(gnb_accuracy)
```

```python
svc=SVC()
svc.fit(input_vector_train,output_label_train)
svc_prediction=svc.predict(input_vector_test)
predict=svc_prediction
print(predict)
svc_accuracy=accuracy_score(output_label_test,svc_prediction)

print(svc_accuracy)
```

```python
gradientboosting=GradientBoostingClassifier()
gradientboosting.fit(input_vector_train,output_label_train)
gradientboosting_prediction=gradientboosting.predict(input_vector_test)
predict=gradientboosting_prediction
print(predict)
gb_accuracy=accuracy_score(output_label_test,gradientboosting_prediction)

print(gb_accuracy)
```

```python
baggingclassifier=BaggingClassifier()
baggingclassifier.fit(input_vector_train,output_label_train)
baggingclassifier_prediction=baggingclassifier.predict(input_vector_test)
predict=baggingclassifier_prediction
print(predict)
bc_accuracy=accuracy_score(output_label_test,baggingclassifier_prediction)

print(bc_accuracy)
```

```python
x=PrettyTable()
x.field_names=["Model","Accuracy"]
x.add_row(["LogisticRegression",lr_accuracy.round(2)])
x.add_row(["RandomForestClassifier",rfc_accuracy.round(2)])
x.add_row(["LinearSVC,LogisticRegression",lr_accuracy.round(2)])
x.add_row(["BernoulliNB",lsvc_accuracy.round(2)])
x.add_row(["DecisionTreeClassifier",dt_accuracy.round(2)])
x.add_row(["KNeighborsClassifier",knn_accuracy.round(2)])
x.add_row(["GaussianNB",gnb_accuracy.round(2)])
x.add_row(["SVC",svc_accuracy.round(2)])
x.add_row(["GradientBoostingClassifier",gb_accuracy.round(2)])
x.add_row(["BaggingClassifier",bc_accuracy.round(2)])
print(x)
```
